```python
import torchvision
model_ft = torchvision.models.vgg16(pretrained=True)
```

```python
from PIL import Image
import torchvision.transforms as transforms
import numpy as np 
image = Image.open("/media/shariq/ssd_storage/Udacity-DeepLearning/convolutional-neural-networks/bird.jpg")

preprocess = transforms.Compose([transforms.RandomResizedCrop(224),transforms.ToTensor()])
img_tensor = preprocess(image)
print(img_tensor.shape)
dic ={} 
x = img_tensor
x = img_tensor.unsqueeze(0)
print(x.shape)    
for i,j in model_ft.features._modules.items():
    print(i)
    x = j(x)
    dic[i] =x
    print(i,x.shape)
```

# Visualizing the CNN layers



Here I have visualized the 1st layer of the neural network

```python
import matplotlib.pyplot as plt
feature_maps = dic["1"].detach().numpy()
feature_maps = feature_maps.transpose(0,2,3,1)
print(feature_maps.shape)
square = 8
ix = 1
plt.figure(figsize=(15,15))
for _ in range(square):
  for _ in range(square):
     # specify subplot and turn of axis
     ax = plt.subplot(square, square, ix)
     ax.set_xticks([])
     ax.set_yticks([])
     # plot filter channel in grayscale
     plt.imshow(feature_maps[0, :, :, ix-1],cmap ="gray" )
     ix += 1
# show the figure
plt.show()
```
